```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fn = pd.read_csv("fandango_scrape.csv")
al = pd.read_csv("all_sites_scores.csv")
```

```python
al.info()
```

```python
fn.info()
```

```python
fn[fn["STARS"] > fn["RATING"]]
```

```python
al.head()
```

```python
ls = list(al["FILM"])
```

```python
fn = fn[fn["FILM"].isin(ls)]
fn.info()
```

```python
fn.set_index("FILM")
```

```python
al.set_index("FILM")
```

```python
ch = pd.merge(fn, al, how = "inner")
ch.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot("RATING",color = "red", alpha = 0.4, data = ch)
ax.plot("IMDB",color = "blue", alpha = 0.4, data = ch)
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(x = "RATING", y = "VOTES", data = ch)
plt.xlim(0, 5)
```

```python
sns.heatmap(ch.corr(), cmap = "viridis")
```

```python
ch30 = ch[ch["VOTES"]>30]
ch30["IMDB"] /= 2
ch30["RottenTomatoes"] = 0.05*ch30["RottenTomatoes"]
ch30["RottenTomatoes_User"] = 0.05*ch30["RottenTomatoes_User"]
ch30["Metacritic"] = 0.05*ch30["Metacritic"]
ch30["Metacritic_User"] /= 2
ch30.head(4)
```

```python
sns.countplot(x = "STARS", color = "red", alpha = 0.4, data = ch30)
```

```python
sns.histplot(x = "RATING", color = "red", alpha = 0.4, data = ch30, kde = True)
```

```python
sns.kdeplot(data = ch30, x = "RATING", shade = True, color = "red", clip = [0, 5])
```

```python
fig, axe = plt.subplots(figsize = (10, 8))
axe = sns.histplot(x = "RATING", color = "red", alpha = 0.4, data = ch30, kde = True)
axe = sns.histplot(x = "IMDB", color = "blue", alpha = 0.4, data = ch30, kde = True)
```

```python
plt.figure(figsize = (12, 5), dpi = 150)
sns.kdeplot(ch30["STARS"], color='purple', fill=True, clip = [0, 5], label = "Fandango")
sns.kdeplot(ch30["IMDB"], color='red', fill=True, clip = [0, 5], label = "IMDB")
sns.kdeplot(ch30["RottenTomatoes"], color='blue', fill=True, clip = [0, 5], label = "Rottentomatoes")
sns.kdeplot(ch30["RottenTomatoes_User"], color='yellow', fill=True, clip = [0, 5], label = "Rottentomatoes users")
sns.kdeplot(ch30["Metacritic"], color='orange', fill=True, clip = [0, 5], label = "Metacritic")
sns.kdeplot(ch30["Metacritic_User"], color='brown', fill=True, clip = [0, 5], label = "Metacritic users")
# plt.ylim(0, 0.1)
# yticks = [0, 0.02, 0.04, 0.06, 0.08, 0.1]
# yticklabels = ['0%', '2%', '4%', '6%', '8%', '10%']
plt.legend(loc = 2)
plt.savefig("allComparison.png")
```
